```python


import pandas as pd
import numpy as np
import os

#---------Anaylsis of responses for CONFIDENCE when using ontology user interfaces----------
#Four interfaces include: Wizard, Quick Form, WebProtege, and Wiki Data

df = pd.read_csv('Comparison2_Confidence.csv')
#read Session 2 Comparison Questionnaire CSV file, raw data


#to read column headers use print(comparison2_Confidence.columns)

#df.describe()
#Statistical summary of User-study Session 2 responses from participants after using the software
df.describe(include=[np.object])
       
```

```python
#Confidence- breakdown of level of confidence by study participants after using the interface. 
#Columns broken down for Wizard, QuickFrom, WikiData, and WebProtege
confidence =df[['Confidence_Wizard','Confidence_QuickForm','Confidence_WikiData','Confidence_WebProtege']]

confidence_long=pd.melt(confidence, var_name='Tool', value_name='ConfidenceLevel')
variable_split=confidence_long['Tool'].str.split('_')
variable_split=variable_split.str[1]                              
confidence_long['Tool']=variable_split
confidence=confidence_long[['Tool','ConfidenceLevel']]

#confidence
#
column_values = confidence.columns.values.tolist()
print('Tools & values for confidence levels: ')
print()
for ConfidenceLevel in column_values:
    print(confidence[ConfidenceLevel].unique())

```

```python
#Frequency Table - frequency of responses per ranking category
#Rating of 1 indicates least confidence & 4 is the most helpful
confidence_table=pd.crosstab(confidence_long.ConfidenceLevel, confidence_long.Tool)
confidence_table
```
